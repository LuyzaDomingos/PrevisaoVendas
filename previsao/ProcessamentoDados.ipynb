```python
import pandas as pd
from IPython.display import clear_output

dfs = []
for txt in ['2018.txt', '2019.txt', '2020.txt','2021.txt']:
    df = pd.read_csv(txt, delimiter=';', encoding = 'cp1252')
    df.columns = ['COD_FILIAL', 'SIGLA', 'NOME', 'REGIAO', 'COD_PRODUTO', 'PRODUTO', 'COD_GRUPO', 'CLASSIFICACAO', 'QTD_VENDIDA', 'DATA', 'COD_FILIAL_', 'COD_ENTREGA', 'ENTREGA', 'ANO/MES', 'MES', '1', '2', 'CD', '3', '4', 'SUBCATEGORIA', 'FORNECEDOR']
    df['DATA'] = pd.to_datetime(df['DATA'], dayfirst=True)
    dfs.append(df)
df = pd.concat(dfs) 

df
```

```python
df = df.sort_values(by='DATA', ascending=True)
df.head(2)
```

## Geração da tabela de dados geral

Vendas de produtos por data

```python
dfg = df.drop(columns=['COD_FILIAL', 'SIGLA', 'NOME', 'REGIAO', 'COD_PRODUTO', 'COD_GRUPO', 'CLASSIFICACAO', 'COD_FILIAL_', 'COD_ENTREGA', 'ENTREGA', 'ANO/MES', '1', '2', 'CD', '3', '4', 'SUBCATEGORIA', 'FORNECEDOR'])

index = pd.date_range(start='2018-01-01', end='2021-04-01', freq='D')
columns = dfg['PRODUTO'].unique()

df2 = pd.DataFrame(index=index, columns=columns)
df2.fillna(value=0, inplace=True)
df2.head()
```

```python
p = 0
for product in columns:
    p += 1
    group = dfg.loc[dfg['PRODUTO'] == product].groupby(['DATA']).sum()
    clear_output(wait=True)
    print(p, product)
    i = 0
    for row in group.index:
        df2.loc[row][product] = group.iloc[i]['QTD_VENDIDA']
        i += 1
```

## Geração de dicionários de fornecedores, grupos e classificações



```python
df['SUBCATEGORIA'] = df['SUBCATEGORIA'].apply(lambda x : x.strip())
df['FORNECEDOR'] = df['FORNECEDOR'].apply(lambda x : x.strip())
df['CLASSIFICACAO'] = df['CLASSIFICACAO'].apply(lambda x : x.strip())
```

```python
sub_dict = {}
man_dict = {}
clas_dict = {}
p = 0
for product in columns:
    p += 1
    clear_output(wait=True)
    print(p, product)
    
    subs = df.loc[df['PRODUTO'] == product]['SUBCATEGORIA'].unique()
    mans = df.loc[df['PRODUTO'] == product]['FORNECEDOR'].unique()
    clas = df.loc[df['PRODUTO'] == product]['CLASSIFICACAO'].unique()
    
    for sub in subs:
        if sub in sub_dict:
            sub_dict[sub].add(product)
        else:
            sub_dict[sub] = {product}
    for man in mans:
        if man in man_dict:
            man_dict[man].add(product)
        else:
            man_dict[man] = {product}
    for clas in clas_dict:
        if clas in clas_dict:
            clas_dict[clas].add(product)
        else:
            clas_dict[clas] = {product}
```
